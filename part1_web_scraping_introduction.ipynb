```python
#import the Libraries 
import lxml
from bs4 import BeautifulSoup
import requests
```

```python
# lets pass the HTML script as a string to Beautiful Soup , pass 'html.parser' as the parser
soup = BeautifulSoup('<b class="boldest">Extremely bold</b>', 'html.parser')

# now we can use this soup object to extract element using tags 
tag = soup.b

#ceck the data type of soup.select() element 
print(type(tag))
```

```python
#we can get the element name and attributes using .name and .attars 
tag.name
```

```python
#as tag elements can be treated as a dict object you can get the attribute value using attribute name as the key
tag = BeautifulSoup('<b id="boldest">bold</b>', 'html.parser').b
tag['id']
```

```python
#we can get the element name and attributes using .name and .attars 
tag.attrs
```

     Lets get a new HTML Page with more Tags and CSS 

```python
## passing the entire code as the str object 

htmlstr = \
' <!DOCTYPE html> <html> <head><style>p{color: red;text-align: center;} \
</style></head><body><p src = "testing_attr">Hello Srijeet!</p><p>Welcome \
to Web Scrapping Testing</p></body></html>'
```

```python
#create a soup object using BeautifulSoup and passing 'lxml' as the parser
soup = BeautifulSoup(htmlstr, 'lxml')

type(soup)
```

```python
#lets select the p tag elements from the page
my_information = soup.select('p')

#print the length of selected elements
print(len(my_information))

#printing the type of final result
print('\n',type(my_information))

#printing the types of individual element data types
print('\n',type(my_information[0]))
```

```python
#lets print the entire result
print(my_information)
```

```python
#now lets select an 1st element and print its type
information_1 = my_information[0]
print(information_1)
```

```python
#as this tag element can be treated as dict data type lets get the src attribute value of it 
print(information_1['src'])
```

```python
#Now iy you want the text of tag object you can use either of the below commands 
print(information_1.text)
print(information_1.getText())
```
